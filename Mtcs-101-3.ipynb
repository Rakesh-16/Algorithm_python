```python
#Probabilistic Counting


import random
import numpy as np
import matplotlib.pyplot as plt


def create_fibonacci_sequence(length):
    if length <= 0:
        return []
    elif length == 1:
        return [1]

    fib_seq = [1, 2]

    for _ in range(2, length):
        fib_seq.append(fib_seq[-1] + fib_seq[-2])

    return fib_seq


def biased_coin_flip(lower_bound, upper_bound):
    # Determine the probability of heads
    probability_heads = 1 / (upper_bound - lower_bound)

    # Generate a random float between 0 and 1
    random_value = random.random()

    # Return 1 for heads if random value is less than the probability
    return 1 if random_value < probability_heads else 0


# Number of bits and Fibonacci sequence length
num_bits = 5
fib_length = 32

# Generate a list of powers of 2
power_of_twos = [2 ** i for i in range(20)]
increments_list = []  # To store the number of increments for each simulation

# Run simulations
for _ in range(50000):
    current_index = 0
    counter_list = [0] * num_bits  # Initialize the counter list
    increment_count = 0

    for _ in range(1000):
        if biased_coin_flip(power_of_twos[current_index], power_of_twos[current_index + 1]):
            position = 0
            while counter_list[position] == 1:
                counter_list[position] = 0
                position += 1
            counter_list[position] = 1
            current_index += 1
            increment_count += 1

    increments_list.append(increment_count)

# Calculate average increments
average_increments = np.mean(increments_list)

print(f"Average increments: {average_increments}")

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(range(len(increments_list)), increments_list, label='Number of Increments')
plt.xlabel('Simulation Iteration')
plt.ylabel('Number of Increments')
plt.title('Number of Increments vs. Simulation Iteration')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Dynamic Table
import matplotlib.pyplot as plt


def generate_fibonacci_sequence(length):
    if length <= 0:
        return []
    elif length == 1:
        return [1]

    sequence = [1, 1]

    while len(sequence) < length:
        sequence.append(sequence[-1] + sequence[-2])

    return sequence


# Generate Fibonacci sequence for a length of 10
fibonacci_list = generate_fibonacci_sequence(10)

elements_count = 0  # Current number of elements
current_capacity = 1
fibonacci_index = 1
current_load_factor = 0.0  # (elements_count / current_capacity)
load_factor_history = []
capacity_history = []

for _ in range(10):
    user_input = int(input("Enter 1 to insert or 0 to delete: "))
    if user_input == 1:
        elements_count += 1
        if current_capacity < elements_count:
            current_capacity += fibonacci_list[fibonacci_index]
            fibonacci_index += 1
    else:
        elements_count -= 1
        if elements_count / current_capacity < (fibonacci_list[fibonacci_index - 1] / fibonacci_list[fibonacci_index]):
            current_capacity -= fibonacci_list[fibonacci_index]
            fibonacci_index -= 1

    current_load_factor = elements_count / current_capacity
    load_factor_history.append(current_load_factor)  # Record the load factor
    capacity_history.append(current_capacity)  # Record the capacity


plt.plot(range(len(load_factor_history)), load_factor_history, marker='o', label='Load Factor')
plt.plot(range(len(capacity_history)), capacity_history, marker='x', label='Current Capacity')

plt.xlabel('Stage')
plt.ylabel('Value')
plt.title('Load Factor and Capacity Over Stages')
plt.legend()

plt.show()
```
