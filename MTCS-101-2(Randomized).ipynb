Given a permuted list A containing n numbers in the range [0,n-1],find how many numbers coincide with their index ON AVERAGE.

```python
import random as rnd

def count(permutation):

    count = 0
    for i in range(len(permutation)):
        if permutation[i] == i:
            count += 1
    return count


def experiment(n, trials):

    total_matching = 0
    for _ in range(trials):
        permutation = list(range(n))
        rnd.shuffle(permutation)
        total_matching += count(permutation)

    return total_matching / trials


n = 1000
trials = 10000

matching_points = experiment(n, trials)

print(f"The average numbers conciding with its index for permutations of size {n} is: {matching_points:.2f}")
```

Permute a list of numbers using the following methods:
(i) Permute-by-Sorting
(ii) Randomize in-Place

```python
#Randomly permute IN PLACE
def random_permute(A):
    n = len(A)
    for i in range(n):
        j = random.randint(i,n-1)
        A[i],A[j] = A[j],A[i]

random_permute(A)
print(A)
```

Perform Random Binary Search and compare it with standard Binary Search.

```python
#Randomized Binary Search
def BS(A, X):
    l = 0
    r = len(A) - 1

    while l < r:

        mid = random.randint(l,r)

        if A[mid] < X:
            l = mid + 1
        if A[mid] > X:
            r = mid - 1
        else:
                res = mid
                l = mid+1

    return res
```

```python
#Majority Element

from collections import defaultdict

def take_two_elements_randomly(A):
    n = len(A)
    i = random.randint(0,n-1)
    j = random.randint(0,n-1)
    return A[i],A[j]


def majority_element(A):
    n = len(A)

    counts = defaultdict(int)
    candidate = None
    candidate_count = 0


    for k in range(n):
      i,j = take_two_elements_randomly(A)

      if i == j:             #if two elements are same put one back
        counts[i] += 1
        if counts[i] > candidate_count:
            candidate = i
            candidate_count = counts[i]

      else:
            counts[i] -= 1
            counts[j] -= 1    #if two elements are different remove both
            if counts[i] <= 0:
                del counts[i]
            if counts[j] <= 0:
                del counts[j]
    return candidate

A = [3, 2, 3, 2, 3, 2, 2, 4, 4]
print("Majority element:", majority_element(A))
```

```python
import random

def run_algorithm(n, distribution_type):
    i = n
    count = 0

    while i != 1:
        if distribution_type == 'uniform':
            i = random.randint(1, i)
        elif distribution_type == 'uniform_float':
            i = int(random.uniform(1, i))
        elif distribution_type == 'beta':
            i = int(random.betavariate(1, i))
        elif distribution_type == 'exponential':
            i = int(random.expovariate(1.0))
        elif distribution_type == 'gamma':
            i = int(random.gammavariate(2, i))
        elif distribution_type == 'gaussian':
            i = int(random.gauss(mu=0.0, sigma=1.0))
        else:
            raise ValueError("Unknown distribution type")

        i = max(1, min(i, n))
        count += 1

    return count

def experiment(n_values, distribution_types):
    results = []

    for n in n_values:
        for dist in distribution_types:
            count = run_algorithm(n, dist)
            results.append((n, count, dist))

    return results

n_values = [10, 50, 100, 500, 1000]
distribution_types = ['uniform', 'uniform_float', 'beta', 'exponential', 'gamma', 'gaussian']

results = experiment(n_values, distribution_types)

print(f"{'n':<10} {'count':<10} {'distribution':<15}")
print("-" * 35)
for n, count, dist in results:
    print(f"{n:<10} {count:<10} {dist:<15}")
```

Clock solitaire

```python
import random as rd

def setup_game():
    unique_numbers = rd.sample(range(0, 13), 13) * 4
    rd.shuffle(unique_numbers)

    groups = [[] for _ in range(13)]

    for i, card in enumerate(unique_numbers):
        groups[i % 13].append(card)

    return groups

def ClockSolitaire(groups):
    current_position = 12
    cards_accessed_count = 0

    while cards_accessed_count < 52:
        if not groups[current_position]:
            return False

        card = groups[current_position].pop()
        current_position = card
        cards_accessed_count += 1

    return True  # All cards were accessed

def simulate(n=1000000):
    wins = 0
    for _ in range(n):
        groups = setup_game()
        if ClockSolitaire(groups):
            wins += 1
    return wins / n

print(f"Win rate: {simulate() * 100:.2f}%")
```

```python
import random as rd

def setup_game():
    unique_numbers = rd.sample(range(0, 13), 13) * 4
    rd.shuffle(unique_numbers)

    groups = [[] for _ in range(13)]

    for i, card in enumerate(unique_numbers):
        groups[i % 13].append(card)

    return groups

def ClockSolitaire(groups):
    current_position = 12
    cards_accessed_count = 0

    while True:
        if not groups[current_position]:

            return False

        card = groups[current_position].pop()
        next_position = card % 13

        if cards_accessed_count == 52:

            return True

        current_position = next_position

def simulate(n=1000000):
    wins = 0
    for _ in range(n):
        groups = setup_game()
        if ClockSolitaire(groups):
            wins += 1
    return wins / n


print(f"Win rate: {simulate() * 100:.2f}%")
```

```python
#Flip matrix

import random
from typing import List

class Solution:

    def __init__(self, m: int, n: int):
        self.m = m
        self.n = n
        self.total = m * n
        self.used = set()
        self.r = random.Random()

    def flip(self) -> List[int]:
        r = self.r.randint(0, self.total - 1)
        while r in self.used:
            r = self.r.randint(0, self.total - 1)
        self.used.add(r)
        return [r // self.n, r % self.n]

    def reset(self) -> None:
        self.used.clear()
```
