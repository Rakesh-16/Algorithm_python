```python
def BS(A, X):
    l = 0
    r = len(A) - 1

    while l < r:
        mid = (l + r) // 2

        if A[mid] < X:
            l = mid + 1
        if A[mid] > X:
            r = mid - 1
        else:
                res = mid
                l = mid+1

    return res
```

```python
def fibonacci_recursive(n, count=0):

    count += 1

    if n <= 0:
        return 0, count
    elif n == 1:
        return 1, count
    else:

        fib_n_1, count = fibonacci_recursive(n-1, count)
        fib_n_2, count = fibonacci_recursive(n-2, count)
        return fib_n_1 + fib_n_2, count


n = 5
result, calls = fibonacci_recursive(n)
print(f"The {n}-th Fibonacci number is: {result}")
print(f"Number of function calls made: {calls}")
```

```python

def gcd(m, n):
    while n != 0:
        m, n = n, m % n
    return m


def fibonacci(num):

    if num == 1:
        return 1
    elif num == 2:
        return 1

    fib1, fib2 = 1, 1
    for _ in range(num - 1):
        fib1, fib2 = fib2, fib1 + fib2
    return fib2

def verify_gcd_fibonacci_property(m, n):

    gcd_mn = gcd(m, n)

    fib_m = fibonacci(m)
    fib_n = fibonacci(n)


    gcd_fib_mn = gcd(fib_m, fib_n)


    fib_gcd_mn = fibonacci(gcd_mn)

    # Verify the property
    print(f"GCD(m, n): {gcd_mn}")
    print(f"F_m: {fib_m}")
    print(f"F_n: {fib_n}")
    print(f"GCD(F_m, F_n): {gcd_fib_mn}")
    print(f"F_GCD(m, n): {fib_gcd_mn}")

    if gcd_fib_mn == fib_gcd_mn:
        print("Verification successful: GCD(F_m, F_n) matches F_GCD(m, n).")
    else:
        print("Verification failed: GCD(F_m, F_n) does not match F_GCD(m, n).")


m = int(input("Enter the first number: "))
n = int(input("Enter the second number: "))

verify_gcd_fibonacci_property(m, n)
```

```python
# insertion sort
def insertion(A):
  for i in range(1,len(A)):
    key = A[i]
    j = i-1
    while(j >= 0 and A[j] > key):
      A[j+1] = A[j]
      j = j-1
    A[j+1] = key
  return A

A = [12, 11, 13, 5, 6]
print(insertion(A))
```

```python
# selection sort
def selection(A):
  for i in range(len(A)):
    min_idx = i
    for j in range(i+1, len(A)):
      if A[min_idx] > A[j]:
        min_idx = j
    A[i], A[min_idx] = A[min_idx], A[i]
  return A

A = [12, 11, 13, 5, 6]
print(selection(A))
```

```python
#12 radix sort using queue and by counting sort starting from unit position

def radix_sort(A):
  queue = []
  for i in range(len(A)):
    queue.append(A[i])

  max_val = max(A)
  exp = 1

  while max_val // exp > 0:
    counting_sort_queue(queue, exp)
    exp *= 10

  return queue

def counting_sort_queue(queue, exp):
  count = [0] * 10  # 0-9 digits
  output = [0] * len(queue)

  # Count occurrences of each digit at the current place value
  for i in range(len(queue)):
    index = (queue[i] // exp) % 10
    count[index] += 1


  for i in range(1, 10):
    count[i] += count[i - 1]

  #output queue
  i = len(queue) - 1
  while i >= 0:
    index = (queue[i] // exp) % 10
    output[count[index] - 1] = queue[i]
    count[index] -= 1
    i -= 1

  for i in range(len(queue)):
    queue[i] = output[i]


A = [170, 45, 75, 90, 802, 24, 2, 66]
sorted_A = radix_sort(A)
print(sorted_A)
```

```python
#countingsort
def counting_sort(A ,n ,k):
  B = [0] * n
  C = [0] * (k+1)

  for j in range(n):
    C[A[j]] += 1

  for i in range(1,k+1):
    C[i] += C[i-1]

  for j in range(n-1,-1,-1):
    B[C[A[j]]-1] = A[j]
    C[A[j]] = C[A[j]] - 1

  return B

A = [170, 45, 75, 9, 802, 24, 2, 66]
sorted_A = counting_sort(A,len(A),max(A))
print(sorted_A)
```

```python
#Quick sort
def partition(A, l, r):

    pivot = A[l]
    i = l + 1
    j = r
    while i <= j:
        if A[i] <= pivot:
            i += 1
        elif A[j] > pivot:
            j -= 1
        else:
            A[i], A[j] = A[j], A[i]
            i += 1
            j -= 1
    A[l], A[j] = A[j], A[l]
    return j


def quick_sort(A, l, r):
    if l < r:
        pivot_index = partition(A, l, r)
        quick_sort(A, l, pivot_index - 1)
        quick_sort(A, pivot_index + 1, r)


A = [12, 11, 13, 5, 6]
quick_sort(A, 0, len(A) - 1)
print(A)

#A = [12, 11, 13, 5, 6]
#print(partition(A,0,len(A)-1))
```

```python
#Heapify Bottom up
def Heapify(A, i):
  n = len(A)

  l = (2*i) + 1
  r = (2*i) + 2
  if l < n and A[i] < A[l]:
    largest = l
  else: largest = i
  if r < n and A[largest] < A[r]:
      largest = r
  if largest != i:
    A[i], A[largest] = A[largest], A[i]
    Heapify(A, largest)
  return A



def Buid_Max_Heap(A,n):
  n = len(A)
  for i in range(n//2, -1,-1):
      Heapify(A, i)



A = [12, 11, 13, 5, 6, 8, 14]
Buid_Max_Heap(A,len(A))
print(A)
```

```python
# Top down heapify

def heapify_up(A, i):

    parent = (i - 1) // 2
    while i > 0 and A[i] > A[parent]:
        A[i], A[parent] = A[parent], A[i]
        i = parent
        parent = (i - 1) // 2

def build_max_heap_top_down(A):

    heap = []
    for value in A:
        heap.append(value)
        heapify_up(heap, len(heap) - 1)

    A[:] = heap


A = [12, 11, 13, 5, 6, 8, 14]
build_max_heap_top_down(A)
print(A)
```

```python
#16 > Bucket sort

def insertion_sort(bucket):

    for i in range(1, len(bucket)):
        key = bucket[i]
        j = i - 1
        while j >= 0 and bucket[j] > key:
            bucket[j + 1] = bucket[j]
            j -= 1
        bucket[j + 1] = key

def bucket_sort(A):

    n = len(A)

    B = [[] for _ in range(n)]


    for i in range(n):
        index = int(n * A[i])  #bucket index
        B[index].append(A[i])


    for i in range(n):
        insertion_sort(B[i])


    sorted_A = []
    for i in range(n):
        sorted_A.extend(B[i])

    return sorted_A


if __name__ == "__main__":
    A = [0.78, 0.17, 0.39, 0.26, 0.72]
    sorted_A = bucket_sort(A)
    print("Sorted array:", sorted_A)
```

```python
#17. Determine the 1st,2nd and 3rd largest DISTINCT elements in a list using a O(n) algorithm. (Hint : Use Heap)



def heapify(arr, n, i):

    smallest = i
    left = 2 * i + 1
    right = 2 * i + 2


    if left < n and arr[left] < arr[smallest]:
        smallest = left

    if right < n and arr[right] < arr[smallest]:
        smallest = right


    if smallest != i:
        arr[i], arr[smallest] = arr[smallest], arr[i]
        heapify(arr, n, smallest)

def push_heap(heap, item):

    heap.append(item)
    index = len(heap) - 1
    parent = (index - 1) // 2

    while index > 0 and heap[index] < heap[parent]:
        heap[index], heap[parent] = heap[parent], heap[index]
        index = parent
        parent = (index - 1) // 2

def pop_heap(heap):

    if len(heap) == 0:
        return None

    root = heap[0]
    last_item = heap.pop()

    if len(heap) > 0:
        heap[0] = last_item
        heapify(heap, len(heap), 0)

    return root

def find_top_k_distinct_largest(arr, k):


    distinct_elements = set(arr)

    min_heap = []

    for num in distinct_elements:
        if len(min_heap) < k:
            push_heap(min_heap, num)
        else:
            if num > min_heap[0]:
                pop_heap(min_heap)
                push_heap(min_heap, num)


    return sorted(min_heap, reverse=True)



arr = [10, 20, 4, 110, 30, 30, 50, 60]
top_3_largest = find_top_k_distinct_largest(arr, 3)
print(top_3_largest)
```

```python
#18 > Using the idea of Partition,determine the K-th smallest element in a given list,where K is an input given by the user.


def find_kth_smallest(nums, k):

    return kth_smallest(nums, 0, len(nums) - 1, k)

def three_way_partition(nums, low, high):

    pivot = nums[low]
    lt = low
    gt = high
    i = low + 1

    while i <= gt:
        if nums[i] < pivot:
            nums[lt], nums[i] = nums[i], nums[lt]
            lt += 1
            i += 1
        elif nums[i] > pivot:
            nums[i], nums[gt] = nums[gt], nums[i]
            gt -= 1
        else:
            i += 1

    return lt, gt

def kth_smallest(nums, low, high, k):

    if low == high:
        return nums[low]

    lt, gt = three_way_partition(nums, low, high)

    count = gt - lt + 1

    if k <= (lt - low):
        return kth_smallest(nums, low, lt - 1, k)
    elif k <= (lt - low + count):
        return nums[lt]
    else:
        return kth_smallest(nums, gt + 1, high, k - (lt - low + count))


nums = [3, 2, 1, 5, 6, 4]
k = 2
print(find_kth_smallest(nums, k))
```
